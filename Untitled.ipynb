```python
df = df.drop("Unnamed: 0", axis = 1)
df.head(10)
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5-10","4-5")
```

```python
df.head(30)
```

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data= df,x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

#from the above chart we have analysed that:
#the number of females in the data is more than males data

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
gb
```

```python
plt.figure(figsize=(4,5))
plt.title("Relationship between Parents education and Students Score")
sns.heatmap(gb, annot= True)
plt.show()
```

```python
# from the above chart we have concluded that the education of 
#the parents have a good impact
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
gb1
```

```python
plt.figure(figsize=(5,3))
plt.title("Relationship between Parents education and Students Score")
sns.heatmap(gb1, annot= True)
plt.show()
```
